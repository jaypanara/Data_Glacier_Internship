## **Analysis of the Dataframe**

Plotting the number of pink and yellow cabs running in the city:

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set(font_scale=1.4)
new_Cab_Data['Company'].value_counts().plot(kind='bar', figsize=(7, 6), rot=0);
plt.xlabel("Company", labelpad=14)
plt.ylabel("Number of Cabs", labelpad=14)
plt.title("Number of cabs of a particular company running in the cities", y=1.02);
```

It is observed that there are more cabs of Yellow cab company running around the cities.

```python
plt.figure(figsize=(10, 10));
Users_City=sns.barplot(x='Users', y='City', data=new_Cab_Data);
```

It is observed that the highest number of users are from New York.

```python
pink_cab = new_Cab_Data[new_Cab_Data.Company.isin(["Pink Cab"])]
pink_cab.head()
yellow_cab = new_Cab_Data[new_Cab_Data.Company.isin(["Yellow Cab"])]
yellow_cab.head()
```

```python
pink_transaction = pink_cab.groupby("Year of Travel")['Transaction ID'].count()
yellow_transaction = yellow_cab.groupby("Year of Travel")['Transaction ID'].count()

labels=['2016','2017','2018']
plt.figure(figsize=(17,9))
ax1 = plt.subplot(111)
ax1.bar(labels, pink_transaction.values, width=0.4, color='pink', align='center',label='Pink Cab')
ax1.bar(labels, yellow_transaction.values, width=0.4, color='yellow', align='center',label='Yellow Cab',bottom=pink_transaction)
plt.title("Pink cab vs Yellow cab",fontsize = 20)
plt.ylabel('No. of Transactions',fontsize = 15)
plt.xlabel('Year',fontsize = 15)
plt.legend()
plt.show()
```

It is observed that there are more transaction done for the yellow cab than the pink cabs in three years span.

Let's observe the distribution of various age groups:

```python
plt.figure(figsize=(15,15))
sns.set(style="darkgrid")
sns.histplot(data=new_Cab_Data, x="Age")
plt.show()
```

It is observed that the age groups above 40 do not use the cab services much.

Let's check the relation between price charged and the distance travelled by the two cab companies.

```python
sns.relplot(
    data=new_Cab_Data, kind="line",
    x="KM Travelled", y="Price Charged", hue="Company",
)
```

It is observed that the prices of the yellow cab are higher for each distance travelled when compared to the pink cab.

Let's observe the profit which the companies received from the trips and for that we need to create another column of profits.

```python
new_Cab_Data['Profit'] = new_Cab_Data['Price Charged'] - new_Cab_Data['Cost of Trip']
```

```python
new_Cab_Data.head()
```

```python
import numpy as np
from matplotlib import pyplot as plt
plt.title("Company wise profit")
profit_pie = new_Cab_Data.groupby('Company')['Profit'].sum()
colour = ['pink', 'yellow']
profit_pie.plot(kind='pie',y='Company',startangle=0,figsize=(15,12),autopct='%1.5f%%',explode=[0.03]*2, colors = colour)
```

It is observed that the profit share of yellow cab is more than the pink cab.

Adding another column for the year of travel.

```python
new_Cab_Data['Year of Travel']=pd.DatetimeIndex(new_Cab_Data['Date of Travel']).year
new_Cab_Data.info()
```

Let's observe the yearly profit for botht the companies:

```python
plt.figure(figsize=(10,10))
sns.pointplot(data=new_Cab_Data, x=new_Cab_Data["Year of Travel"], y=new_Cab_Data["Profit"], hue='Company')
plt.title('Yearly Profit',fontsize = 20)
plt.ylabel("Profit",fontsize = 16)
plt.xlabel("Year",fontsize = 16)
plt.show
```

It is seen from the graph that the yearly profit dropped from 2017 to 2018.

Let's observe if there is any relation between distance travelled and the profit.

```python
plt.figure(figsize=(20,10))
sns.lineplot(data=new_Cab_Data, x='KM Travelled', y='Profit', hue='Company')
plt.title('KM Travelled vs Profit',fontsize = 20)
plt.ylabel('Profit',fontsize = 16)
plt.xlabel('KM Travelled',fontsize = 16)
plt.show()
```

It is seen that more profit is earned as the KMs increases for both the companies but the profit earned by Yellow cabs is more compared to the pink cabs.

Lets observe the realtion between all the features in the dataframe:

```python
sns.pairplot(new_Cab_Data)
```

Let's observe if there are any outliers for any feature:

```python
fig,axes = plt.subplots(3,2 , figsize=(18, 18))
fig.suptitle('Boxplot graph')
sns.boxplot(ax=axes[0,0], x='KM Travelled', data=new_Cab_Data, y="Company", hue="Company", dodge=False)
sns.boxplot(ax=axes[0,1], x='Price Charged', data=new_Cab_Data, y="Company", hue="Company", dodge=False) 
sns.boxplot(ax=axes[1,0], x='Cost of Trip', data=new_Cab_Data, y="Company", hue="Company", dodge=False)
sns.boxplot(ax=axes[1,1], x='Income (USD/Month)', data=new_Cab_Data, y="Company", hue="Company", dodge=False)
sns.boxplot(ax=axes[2,0], x='Profit', data=new_Cab_Data, y="Company", hue="Company", dodge=False)
sns.boxplot(ax=axes[2,1], x='Age', data=new_Cab_Data, y="Company", hue="Company", dodge=False)
```

It is obeserved that there are outliers present in prices charged by the companies and the profits each company received from the trips.
